```python
import matplotlib.pyplot as plt
import os
import cv2
img_path = "./data/train/Closed_Eyes/s0001_01476_0_1_0_0_0_01.png"

img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img_array, cmap="gray")
plt.show()
```

```python
import numpy as np
from scipy.signal import convolve2d

img96 = np.zeros((96, 96), dtype=int)
img96[1:95, 1:95] = img_array


print(img96.shape)
print(img96)
kernel = np.array([[0, 5, 0], [0, 5, 0], [0, 5, 0]])
kernel2 = np.array([[0, 0, 0], [5, 5, 5], [0, 0, 0]])
kernel3 = np.array([[0, 0, 5], [0, 5, 0], [5, 0, 0]])

kernel4 = np.array([[0, 0, 0], 
                   [0, 10, 0], 
                   [0, 0, 0]])

after_kernel = convolve2d(img96, kernel4, mode='same', boundary='fill', fillvalue=0)
```

```python
plt.imshow(after_kernel, cmap="gray")
plt.show()
```
